```python
from keras.datasets import boston_housing
```

```python
import numpy as np
```

```python
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()

mean = train_data.mean(axis = 0)
train_data -= mean
std = train_data.std(axis = 0)
train_data /= std

test_data -= mean
test_data /= std

print("TrainData ColumnMin:",train_data.min(0))
print("TrainData ColumnMax:",train_data.max(0))
print("TrainData ColumnMean:",train_data.mean(0))
print("TrainData ColumnStd:",train_data.std(0))
```

K-Fold-Cross-Validation: 
Let k=4, split the Data into 4 parts 101 data points each

```python
from keras import models, layers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(
        64,
        activation = "relu",
        input_shape = (train_data.shape[1],)
    ))
    model.add(layers.Dense(64, activation = "relu"))
    model.add(layers.Dense(1))
    model.compile(
        optimizer = "rmsprop",
        loss = "mse",
        metrics=["mae"]
    )
    return model
    
model = build_model()
model.summary()
```

```python
k = 4
num_val_samples = len(train_data) // k
num_epochs = 50 # TODO: change to 100
all_scores = []
all_histories = []
all_models = []
all_scores_on_test = []


for i in range(k):
    print("Fold:",str(i+1))
    
    val_data = train_data[ i*num_val_samples : (i+1)*num_val_samples ]
    val_targets = train_targets[ i*num_val_samples : (i+1)*num_val_samples ]
    
    train_data_left = train_data[ :(i*num_val_samples)]
    train_data_right = train_data[(i+1)*num_val_samples : ]
    
    train_targets_left = train_targets[ :(i*num_val_samples)]
    train_targets_right = train_targets[(i+1)*num_val_samples : ]
    
    partial_train_data = np.concatenate(
        [train_data_left, train_data_right], 
        axis = 0
    )
    partial_train_targets = np.concatenate(
        [train_targets_left, train_targets_right], 
        axis = 0
    )
    
    '''
    print("ValData:",val_data.shape)
    print("ValTargets:",val_targets.shape)
    print("TrainLeft:",train_data_left.shape)
    print("TrainRight:",train_data_right.shape)
    print("TargetsLeft:",train_targets_left.shape)
    print("TargetsRight:",train_targets_right.shape)
    print("PartialTrain:",partial_train_data.shape)
    print("PartialTargets:",partial_train_targets.shape)
    '''
    
    model = build_model()
    history = model.fit(
        partial_train_data, partial_train_targets,
        epochs = num_epochs,
        batch_size = 1,
        verbose = 0,
        validation_data=(val_data,val_targets)
    )
    model.save("05_regression_" + str(i+1) + "-model.h5")
    val_mse, val_mae = model.evaluate(val_data, val_targets)
    test_mse, test_mae = model.evaluate(test_data, test_targets)
    all_scores.append(val_mae)
    all_scores_on_test.append(test_mae)
    all_histories.append(history)
    all_models.append(model)
```

```python
import matplotlib.pyplot as plt

print("available keys:",history.history.keys())

start_idx = 10

for key in sorted(history.history.keys()):
    plt.figure(figsize=(20,5))
    for i in range(k):
        plt.plot(all_histories[i].history[key][start_idx:], label= key + " " + str(i+1))
    plt.legend()
    plt.show()
    plt.close()

plt.plot(all_scores, label = "final scores on validation")
plt.plot(all_scores_on_test, label = "final scores on test")
plt.legend()
plt.show()
plt.close()
```
