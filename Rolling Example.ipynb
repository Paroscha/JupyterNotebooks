```python
import numpy as np
import matplotlib as mlb
import pandas as pd
```

```python
X = np.random.rand(100,100)
print(type(X))
```

```python
x = np.arange(10)
x = np.repeat(x,10)
x
```

```python
X[:,0] = x
X[0,:] = x
```

```python
X
```

```python
df = pd.DataFrame(X)
df
```

```python
df.rolling(window = 10, axis = 1).mean()
```

```python
df.rolling(window = 10, axis = 1).mean().iloc[:,range(9,100,10)]
```
